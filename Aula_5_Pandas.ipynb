# **Trabalhando com datas**

```python
#Transformando a coluna de data em tipo inteiro 
df['Data'] = df['Data'].astype("int64")
```

```python
#Verificandn o tipo de dado de cada coluna
df.dtypes
```

```python
#Trasnformando a coluna de data em data
df['Data'] = pd.to_datetime(df['Data'])
```

```python
df.dtypes
```

```python
#Agrupando por ano
df.groupby(df['Data'].dt.year)['Receita'].sum()
```

```python
#Criando uma nova coluna com o ano
df['Ano_Venda'] = df['Data'].dt.year
```

```python
df.sample(5)
```

```python
#Extraindo o mes e o dia
df['Mes_Venda'], df['Dia_Venda'] = (df['Data'].dt.month, df['Data'].dt.day)
```

```python
df.sample(5)
```

```python
#Mostrando a data mais antiga
df['Data'].min()
```

```python
#Calculando a difern√ßa de dias
df['Diferenca_dias'] = df['Data'] - df['Data'].min()
```

```python
df.sample(5)
```

```python
#Criando a coluna de trimestre
df['Trimestre_venda'] = df['Data'].dt.quarter
```

```python
df.sample(5)
```

```python
#Filtrando as vendas de 2019 do mes de marco
vendas_marco_2019 = df.loc[(df['Data'].dt.year == 2019) & (df['Data'].dt.month == 3)]
```

```python
vendas_marco_2019.sample(20)
```
